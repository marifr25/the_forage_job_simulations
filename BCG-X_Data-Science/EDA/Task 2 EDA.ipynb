- margin_net_pow_ele = net margin on power subscription

- net_margin = total net margin
- num_years_antig = antiquity of the client (in number of years)
- origin_up = code of the electricity campaign the customer first subscribed to
- pow_max = subscribed power
- churn = has the client churned over the next 3 months

## Plotting Data

```python
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

%matplotlib inline
```

### client_data.csv

##### Overall Churn vs. Non-Churn Customers

```python
churn_counts = client_df['churn'].value_counts()
```

```python
plt.pie(churn_counts, labels=['Non-Churn', 'Churn'], autopct='%1.1f%%', startangle=140)
plt.title('Overall Churn vs. Non-Churn Customers')
plt.show()
```

##### Churn Rate by `channel_sales`

```python
# Create a mapping for channel_sales'
unique_channels = [channel for channel in client_df['channel_sales'].unique()]
channel_mapping = {channel: f'channel {i+1}' for i, channel in enumerate(unique_channels)}

# Apply the mapping
client_df['channel_sales_mapped'] = client_df['channel_sales'].map(channel_mapping)

# Calculate churn rate by mapped channel_sales
churn_rate = client_df.groupby('channel_sales_mapped')['churn'].mean().reset_index()
```

```python
ax = sns.barplot(x='channel_sales_mapped', y='churn', data=churn_rate, palette='viridis')
plt.xticks(rotation=45, ha='right')
plt.title('Churn Rate by Sales Channel')
plt.xlabel('Sales Channel')
plt.ylabel('Churn Rate')
plt.grid(linestyle='--', linewidth=1, axis='y', alpha=0.75)
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```

#### Churn Rate by Electricity Campaign

```python
unique_campaign = [campaign for campaign in client_df['origin_up'].unique()]
campaign_mapping = {campaign: f'campaign {i+1}' for i, campaign in enumerate(unique_campaign)}
client_df['campaign_mapped'] = client_df['origin_up'].map(campaign_mapping)
campaign_churn_rate = client_df.groupby('campaign_mapped')['churn'].mean().reset_index()
```

```python
ax = sns.barplot(x='campaign_mapped', y='churn', data=campaign_churn_rate, palette='viridis')
plt.title('Churn Rate by Electricity Campaign')
plt.xlabel('Electricity Campaign')
plt.ylabel('Churn Rate')
plt.grid(linestyle='--', linewidth=1, axis='y', alpha=0.75)
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```

#### Churn Rate which is a Gas Client too

```python
client_df['has_gas'] = client_df['has_gas'].replace({'t': 'true', 'f': 'false'})
client_df['has_gas'] = client_df['has_gas'].astype('category')
gasclient_churnrate = client_df.groupby('has_gas')['churn'].mean().reset_index()
```

```python
ax = sns.barplot(x='has_gas', y='churn', data=gasclient_churnrate, palette='viridis')
plt.title('Churn Rate which is a Gas Client too')
plt.xlabel('Gas Client')
plt.ylabel('Churn Rate')
plt.grid(linestyle='--', linewidth=1, axis='y', alpha=0.75)
for container in ax.containers:
    ax.bar_label(container)
plt.show()
```

#### Client Churn Analysis of Consumption

```python
# Renaming the values in 'churn', convert to category dtype column and create a new column
client_df['category_churn'] = client_df['churn'].replace({0:'non-churn', 
                                                          1: 'churn'}).astype('category')

# Grouping data by 'category_churn' and calculating mean consumption
grouped_consumption = client_df.groupby('category_churn')[['cons_12m', 
                                                           'cons_gas_12m', 
                                                           'cons_last_month']].mean().reset_index()

# Melting the data
consumption_melted = pd.melt(grouped_consumption, 
                             id_vars='category_churn', 
                             value_vars=['cons_12m', 'cons_gas_12m', 'cons_last_month'], 
                             var_name='Consumption Type', 
                             value_name='Average Consumption')

# Renaming the values in 'Consumption Type' column
consumption_melted['Consumption Type'] = consumption_melted['Consumption Type'].replace({'cons_12m': 'Electricity Consumption', 
                                                                                         'cons_gas_12m': 'Gas Consumption', 
                                                                                         'cons_last_month': 'Electricity Consumption Last Month'})
```

```python
# Plotting using Seaborn
ax = sns.barplot(x='category_churn', 
            y='Average Consumption', 
            hue='Consumption Type', 
            data=consumption_melted, 
            palette='viridis')

# Set labels and title
plt.xlabel('Category Churn')
plt.ylabel('Average Consumption')
plt.title('Client Churn Analysis of Consumption')
plt.grid(linestyle='--', linewidth=1, axis='y', alpha=0.75)
for container in ax.containers:
    ax.bar_label(container)

# Add legend
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Show plot
plt.show()
```

#### Client Churn Analysis of Forecast

##### Forecast Electricity Consumption

```python
grouped_forecast = client_df.groupby('category_churn')[['forecast_cons_12m',
                                                        'forecast_cons_year']].mean().reset_index()

forecast_melted = pd.melt(grouped_forecast, 
                             id_vars='category_churn', 
                             value_vars=['forecast_cons_12m','forecast_cons_year'], 
                             var_name='Forecast Type', 
                             value_name='Average Forecast')

forecast_melted['Forecast Type'] = forecast_melted['Forecast Type'].replace({'forecast_cons_12m': 'Forecast Consumption',
                                                                             'forecast_cons_year': 'Forecast Consumption (Calendar)'})
```

```python
# Plotting using Seaborn
ax = sns.barplot(x='category_churn', 
                 y='Average Forecast', 
                 hue='Forecast Type', 
                 data=forecast_melted, 
                 palette='viridis')

# Set labels and title
plt.xlabel('Category Churn')
plt.ylabel('Average Forecast')
plt.title('Client Churn of Forecast Electricity Consumption')
plt.grid(linestyle='--', linewidth=1, axis='y', alpha=0.75)
for container in ax.containers:
    ax.bar_label(container)

# Add legend
plt.legend(loc='upper left', bbox_to_anchor=(1, 1))

# Show plot
plt.show()
```
